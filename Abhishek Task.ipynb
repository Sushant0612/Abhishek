```python
import pandas as pd
import seaborn as sns
import datetime
df=pd.read_csv(r"C:\Users\OPKE User\Desktop\abhishek\dataset_1.csv")
df['Session date']=pd.to_datetime(df['Session date'])
display(df.head()),df.info()
```

```python
df['Session type'].unique()
```

```python
# 1> Distribution of users by number of sessions
df['User ID'].value_counts()
```

```python
after7=(min(df['Session date']) + datetime.timedelta(days=7)).strftime("%Y-%m-%d")
after7
```

```python
# >2.b Which session type (A, B, or C?) is able to bring back users most often
# within a period of 7 days since the first session?
result=df[(df['Session date']>=min(df['Session date']))& (df['Session date']<=after7)&(df['Session number']!=1)]
result
```

```python
result['Session type'].value_counts()
# after this result we can say that type c brings back more user within 7 days interval period
```

```python
# Q3>How would you classify the users based on the data analyzed so far?
print(f"Overall Revenue generted by India :{sum(df[df['Country']=='India']['Amount paid (USD)'])}")
print(f"Overall Revenue generted by Session A in india :{sum(df[(df['Country']=='India') & (df['Session type']=='A')]['Amount paid (USD)'])}")
print(f"Overall Revenue generted by Session B in india :{sum(df[(df['Country']=='India') & (df['Session type']=='B')]['Amount paid (USD)'])}")
print(f"Overall Revenue generted by Session C in india :{sum(df[(df['Country']=='India') & (df['Session type']=='C')]['Amount paid (USD)'])}")
```
